```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the csv file from '../data/logs/pix.csv' using pandas
df = pd.read_csv('../data/logs/pix.csv')

# Extract the columns of epoch, loss, and fid from the dataframe
epoch = df['epoch']
loss = df['loss']
fid = df['fid']

# Create a figure and two subplots, one for loss and one for fid
fig, (ax1, ax2) = plt.subplots(2, 1, sharex=True)

# Plot the loss curve on the first subplot
ax1.plot(epoch, loss, color='blue', label='loss')
ax1.set_ylabel('Loss')
ax1.legend()

# Plot the fid curve on the second subplot
ax2.plot(epoch, fid, color='red', label='fid')
ax2.set_xlabel('Epoch')
ax2.set_ylabel('FID')
ax2.legend()

# Show the figure
plt.show()
```
