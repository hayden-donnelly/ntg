```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
noise_dim = 100
img_rows = 256
img_cols = 256
channels = 1
```

```python
fcngan2 = keras.models.load_model('data/saved_models/fcngan_generator2_256x256_e1_bs16.h5')
```

```python
noise = np.random.normal(0, 1, size=(9, noise_dim))

def show_images(noise, generator):
    generated_images = generator.predict(noise)
    plt.figure(figsize=(20, 20))
    
    for i, image in enumerate(generated_images):
        plt.subplot(10, 10, i+1)
        if channels == 1:
            plt.imshow(image.reshape((img_rows, img_cols)), cmap='gray')
        else:
            plt.imshow(image.reshape((img_rows, img_cols, channels)))
        plt.axis('off')
    
    plt.tight_layout()
    plt.show()
    
show_images(noise, generator = fcngan2)
```

```python
noise_index = 4
noise_string = 'public float[] LatentVector1 = {'
for i in range(len(noise[noise_index])):
    noise_string += str(noise[noise_index][i])
    if(i == len(noise[noise_index]) - 1):
        noise_string += 'f'
    else:
        noise_string += 'f, '
    
noise_string += '};'
print(noise_string)
```
